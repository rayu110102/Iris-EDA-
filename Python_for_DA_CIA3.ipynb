# Data Visualization

Correlation Matrix

```python
# CORRELATION OF THE DATASET
sns.pairplot(df) 
```

Scatter Plot

```python
# scatter plot of sepal length, and petal length of all the three species in  iris dataset
sns.FacetGrid(df, hue = 'species',
               height = 6).map (plt.scatter,
                                'sepal_length',
                                'petal_length').add_legend() 
```

```python
sns.FacetGrid(df, hue = 'species',
               height = 6).map (plt.scatter,
                                'sepal_length',
                                'petal_width').add_legend() 
```

```python
sns.FacetGrid(df, hue = 'species',
               height = 6).map (plt.scatter,
                                'sepal_length',
                                'sepal_width').add_legend() 
```

```python
sns.FacetGrid(df, hue = 'species',
               height = 6).map (plt.scatter,
                                'petal_length',
                                'petal_width').add_legend() 
```

Histogram

```python
sns.histplot(df['sepal_length'], edgecolor = 'black', color = 'orange').set(title = "Histplot of Sepal Length")
```

```python
sns.histplot(df['sepal_width'], edgecolor = 'black', color = 'green').set(title = "Histplot of Sepal Width")
```

```python
sns.histplot(df['petal_length'], edgecolor = 'black', color = 'yellow').set(title = "Histplot of Petal Length")
```

```python
sns.histplot(df['petal_width'], edgecolor = 'black', color = 'blue').set(title = "Histplot of Petal Width")
```

Box Plot

```python
sns.boxplot(df['petal_length']).set(title = "boxplot of Petal Length")
```

```python
sns.boxplot(df['petal_width']).set(title = "boxplot of Petal Width")
```
